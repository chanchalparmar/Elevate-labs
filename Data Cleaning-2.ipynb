```python
import pandas as pd
df=pd.read_csv("mpg-data.csv")
```

```python
df
```

```python
df1=df.isna().sum().to_frame()
```

```python
df1
```

```python
df1.rename(columns={0:'Null'},inplace=True)
```

```python
df1
```

```python
df2=df.notna().sum().to_frame()
```

```python
df2.rename(columns={0:'Not Null'},inplace=True)
```

```python
df2
```

```python
df3=pd.concat([df1,df2],axis=1)
df3.reset_index(inplace=True)
```

```python
df3
```

```python
df3.rename({'category':'Category'},axis=1,inplace=True)
df3
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(df3['Category'],df3['Null'] )
plt.xticks(rotation=45)
plt.xlabel('category')
plt.ylabel('No.of Null values')
plt.title('No. of Null values against category')
```

```python
plt.bar(df3['Category'],df3['Not Null'] )
plt.xticks(rotation=45)
```
